```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
area_file_df = pd.read_csv('/Users/soonam/workspace/MLAlgos/area_price.csv')
```

```python
plt.xlabel('area in sqr ft')
plt.ylabel('area US$')
plt.scatter(area_file_df.area,area_file_df.price,color='red',marker='')

reg = linear_model.LinearRegression()
#area_file_df[['area']]
reg.fit(area_file_df[['area']],area_file_df.price)
plt.plot(area_file_df.area,reg.predict(area_file_df[['area']]))
```

```python
#predict for expected value y = mx + b
print(reg.predict(3300))

# Should be same as above prediction Y = reg.coef_ * 3300 + reg.intercept_
y = (reg.coef_ * 3300 ) + reg.intercept_
print(y)
```
